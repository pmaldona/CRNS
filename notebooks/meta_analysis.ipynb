# Meta-Analysis of the response of reaction networks to different perturbation sizes

This notebook demonstrates various plots that help with the analysis of the responses of reaction networks to different perturbation sizes in order to illucidate relationships between structural properties of the networks and their resilience. Most of them are generated using functions from the meta_analysis.py-file, which allows more adaptability.

## 1. Single networks, varying maximum perturbation size

### 1.1. Loading of dataframes

DataFrames for a specific network can be loaded from their folder by the functions
- **get_abstractions_dfs(path, min_max_pert_size=1, max_max_pert_size=0)**
- **get_transitions_dfs(path, min_max_pert_size=1, max_max_pert_size=0)**

Both share the same parameters:
- path (path of the folder in the directory)
- min_pert_size and max_pert_size can be used to make restrictions to the maximum perturbation. Defaults to no restrictions.

```python
from meta_analysis import get_abstractions_dfs, get_transitions_dfs

path = r'C:\Users\simon\Downloads\Bassi\Type_4\Ns30\RNG_Bassi_type_4_N30_Org_6_Id_3'
path = r'C:\Users\simon\Downloads\Bassi\Type_5\Ns40\RNG_Bassi_type_5_N40_Org_10_Id_4'

abstractions_dfs = get_abstractions_dfs(path)
transitions_dfs  = get_transitions_dfs(path)
for a in abstractions_dfs:
    display(a)
```

```python
from numpy import array
resiliences = array(sorted(abstractions_dfs[2]['global_resilience'].tolist()))
x = list(range(len(resiliences)+1))
y = [1]
s = 0
for i, r in enumerate(resiliences):
    s += r
    y.append(1-r)
print(resiliences)

fig, axes = pyplot.subplots()
axes.plot(x,y, lw=3)
axes.tick_params(labelsize=20)
axes.set_xlabel('Number of excluded organizations', fontsize=20)
axes.set_ylabel('Collective resilience\n (sum of global resiliences of included organizations)', fontsize=20)
```

### 1.1. Organizations

```python
from meta_analysis import plot_local_resilience, plot_global_resilience, plot_local_resilience_vs_global, plot_org_size_vs_resilience

colors = generate_random_colors(len(abstractions_dfs[0]))
fig, axes = pyplot.subplots(2,3, figsize=(18,12))
pyplot.suptitle(f'Response of abstractions in {basename(path)}\nto maximum perturbation size', fontsize=25)
pyplot.subplots_adjust(top=0.85, hspace=0.2, wspace=0.25)

plot_local_resilience(axes[0][0], abstractions_dfs, colors)
plot_global_resilience(axes[0][1], abstractions_dfs, colors)
plot_local_resilience_vs_global(axes[0][2], abstractions_dfs, colors)
plot_org_size_vs_resilience(axes[1][0], abstractions_dfs, ['#ff0000', '#00ff00', '#0000ff'], type='local')
plot_org_size_vs_resilience(axes[1][1], abstractions_dfs, ['#ff0000', '#00ff00', '#0000ff'], type='global')

pyplot.show()
```

### 1.2. Transitions

```python
from meta_analysis import plot_number_of_transitions, plot_transition_probabilities, plot_transition_size_vs_probabiliy

colors = generate_random_colors(len(transitions_dfs[0]))
fig, axes = pyplot.subplots(1,3,figsize=(24,8))
pyplot.suptitle(f'Response of transitions in {basename(path)}\n to increasing maximum perturbation size', fontsize=25)
plot_number_of_transitions(axes[0], transitions_dfs)
plot_transition_probabilities(axes[1], transitions_dfs, colors)
plot_transition_size_vs_probabiliy(axes[2], transitions_dfs, colors)
pyplot.subplots_adjust(top=0.8, wspace=0.25)
pyplot.show()
```

```python
import numpy

def histogram(axes, dataframe, column_name: str, bins: int, exclude_first=False):
    distribution = [len(dataframe[(dataframe[column_name] >= i) & (dataframe[column_name] < i+(1/bins))]) for i in numpy.linspace(0,1,bins)]
    distribution = [x/max(distribution) for x in distribution]
    if exclude_first:
        distribution = distribution[1:]
    axes.bar(range(len(distribution)), distribution)
    axes.set_xticks([])
    axes.set_xlabel('global resilience', fontsize=15)
    axes.set_ylabel('frequencies', fontsize=15)
    axes.tick_params(labelsize=15)

column_name = 'global_resilience'
bins = 50

dataframes    = [concat(get_abstractions_dfs_for_multiple_networks(r'C:\Users\simon\Downloads\Bassi\Type_'+type+r'\Ns40',2), ignore_index=True) for type in ['1','2','3','4','5']]
dataframes.append(concat(get_abstractions_dfs_for_multiple_networks(r'C:\Users\simon\Downloads\Rand_Net_markov\Ns40', 2), ignore_index=True))

fig, axes = pyplot.subplots(2,3, figsize=(15,10))

fig.suptitle('Weighted frequencies of global resiliences', fontsize=25)
axes[0][0].set_title('old generator', fontsize=20)
histogram(axes[0][0], dataframes[5], column_name, bins, exclude_first=True)
axes[0][1].set_title('type 1', fontsize=20)
histogram(axes[0][1], dataframes[0], column_name, bins, exclude_first=True)
axes[0][2].set_title('type 2', fontsize=20)
histogram(axes[0][2], dataframes[1], column_name, bins, exclude_first=True)
axes[1][0].set_title('type 3', fontsize=20)
histogram(axes[1][0], dataframes[2], column_name, bins, exclude_first=True)
axes[1][1].set_title('type 4', fontsize=20)
histogram(axes[1][1], dataframes[3], column_name, bins, exclude_first=True)
axes[1][2].set_title('type 5', fontsize=20)
histogram(axes[1][2], dataframes[4], column_name, bins, exclude_first=True)

pyplot.subplots_adjust(wspace=0.3)
pyplot.show()
    
```
